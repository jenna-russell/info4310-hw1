```python
# Simple post-processing script to create a clean subset of tree data
#     - JRz, for personal use and not indended for real world application

#  NOTE: This is eliminating valid, useful data in lieu of something easier to visualize
#        It's possible that this will mask trends, as it is often VERY likely that missing data are non-randomly distributed (e.g. more missing data pre-1955 in this dataset)


import csv
import numpy as np
import pandas as pd

def passes_filter(row):
    # Filter criteria:
    #   Only listed trees that have full species name supplied
#     if len(row['qSpecies']) < 1 or 'Tree(s) ::' in row['qSpecies'] or row['qSpecies'] == "::":
#         return False
    if row['PlantDate'] == "":
        return False
    #   Only trees that have a lat and lng provided
    # elif len(row['Latitude']) < 1 or len(row['Longitude']) < 1:
    #     return False
    #   Only trees with valid SiteInfo
    #elif len(row['qSiteInfo']) < 1 or row['qSiteInfo'] == ':':
        #return False
    #   Only trees that have a DBH provided
    #elif len(row['DBH']) < 1:
        #return False
    else:
        return True

data = []
header = []
with open('Street_Tree_List.csv','r') as f:
    reader = csv.DictReader(f)
    
    header = reader.fieldnames
    for row in reader:
        if passes_filter(row):
            data.append(row)

print(len(data))
```

```python
for a in data: 
    del a['PlantType']
    del a['qCareAssistant']
    del a['PermitNotes']
    del a['XCoord']
    del a['YCoord']
    del a['Location']
    del a['Fire Prevention Districts']
    del a['Police Districts']
    del a['Supervisor Districts']
    del a['Zip Codes']
    del a['Neighborhoods (old)']
```

```python
for d in data:
    date = d["PlantDate"]
    year = date[6:10]
    d["yearPlanted"] = int(year)
```

```python
header = ['TreeID',
 'qLegalStatus',
 'qSpecies',
 'qAddress',
 'SiteOrder',
 'qSiteInfo',
 'qCaretaker',
 'PlantDate',
 'DBH',
 'PlotSize',
 'Latitude',
 'Longitude',
    "yearPlanted"]
```

```python
yearDict = {}
i = 1970
while i < 2020:
    yearDict[i] = 0
    i = i+1
```

```python
data = [i for i in data if i["yearPlanted"] > 1969 and i["yearPlanted"] < 2020]
```

```python
for d in data: 
    d["yearPlanted"] = int(d["yearPlanted"])
    yearDict[d["yearPlanted"]] += 1
```

```python
treePlanted = pd.DataFrame(list(yearDict.items()),columns = ['Year','Trees'])
treePlanted
```

```python
with open('Street_Tree_List-Jenna_FILTERED.csv','w') as f:
    writer = csv.DictWriter(f, fieldnames=header)
    writer.writeheader()
    writer.writerows(data)
```
